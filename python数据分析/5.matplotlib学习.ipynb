```python
# 折线图 plot
# 导入工具箱
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体

rcParams['font.sans-serif'] = ['Heiti TC']

# plt.rcParams['axes.unicode_minus'] = False      # 解决负号显示问题
# 检查当前字体设置
# print("当前字体列表:", plt.rcParams['font.sans-serif'])

# rcParams['font.family'] = ['SimHei']
# rcParams['font.family'] = ['STHeiti']
# 创建一个图表，设置大小
plt.figure(figsize=(10, 5))

# 数据
# month = ['1月的销售数据的销售数据','2月的销售数据的销售数据','3月的销售数据的销售数据','4月的销售数据的销售数据']
month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]

# 绘制折线图
plt.plot(month, sales, label='产品A', linewidth=2, linestyle='--', marker='o', markersize=10)

# 添加标题(通用配置)
plt.title('2025年销售趋势', color='red', fontsize=20)
# 坐标轴的标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额（万元）', fontsize=10)
# 添加图例
plt.legend(loc='upper left')
# 添加网格线
# alpha:设置透明度，color设置颜色，linestyle设置网格线类型：实线；虚线
plt.grid(True, alpha=0.3, color='blue', linestyle='--')
# plt.grid(axis='x') # 只有和x垂直的格子
# plt.grid(axis='y') # 只有和y垂直的格子

# 设置刻度大小
# plt.xticks(rotation = 60 ,fontsize=15)  控制旋转的角度
plt.xticks(rotation=0, fontsize=15)
plt.yticks(rotation=0, fontsize=10)

# 设置y轴的范围
plt.ylim(0, 160)

# 在每个数据点上显示数值
for x, y in zip(month, sales):
    print(x, y)
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=10)

# 显示图标
plt.show()
```

```python
# 绘制饼图：看比例划分
# 导入工具箱
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体

rcParams['font.sans-serif'] = ['Heiti TC']

# 创建一个图表，设置大小
plt.figure(figsize=(10, 5))
# 要绘制的数据
# 逆时针去绘制的饼图
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
# colors = ['red', 'green', 'blue', 'yellow']
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff4498'] # 配色

# 绘制饼图
# startangle:调整初始画图的角度
# autopct:显示百分比
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=90,
        colors=colors,
         )

# 添加标题(通用配置)
plt.title('一天的时间分布', color='red', fontsize=20)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

```python
# 绘制环形图
# 导入工具箱
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体

rcParams['font.sans-serif'] = ['Heiti TC']

# 创建一个图表，设置大小
plt.figure(figsize=(10, 5))
# 要绘制的数据
# 逆时针去绘制的饼图
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
# colors = ['red', 'green', 'blue', 'yellow']
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff4498','yellow'] # 配色

# 绘制饼图
# startangle:调整初始画图的角度
# autopct:显示百分比
# wedgeprops:设置圆环的位置
# pctdistance:设置百分比的位置
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops={'width':0.6},
        pctdistance=0.6
         )

# 添加标题(通用配置)
plt.title('一天的时间分布', color='red', fontsize=20)
plt.text(0,0,'总计:\n100%',ha='center',va='bottom',fontsize=10)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

```python
# 绘制爆炸式饼图
# 导入工具箱
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体

rcParams['font.sans-serif'] = ['Heiti TC']

# 创建一个图表，设置大小
plt.figure(figsize=(10, 5))
# 要绘制的数据
# 逆时针去绘制的饼图
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff4498','yellow'] # 配色
explode = [0.1,0,0,0,0] #设置突出块的位置

# 绘制饼图
# startangle:调整初始画图的角度
# autopct:显示百分比
plt.pie(times, labels=things,
        autopct='%.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        shadow=True,
         )

# 添加标题(通用配置)
plt.title('一天的时间分布', color='red', fontsize=20)
plt.text(0,0,'总计:\n100%',ha='center',va='bottom',fontsize=10)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

```python
# 绘制散点图:数据之前的聚合关系
# 导入工具箱
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
import random

rcParams['font.sans-serif'] = ['Heiti TC']

# 创建一个图表，设置大小
plt.figure(figsize=(10, 8))
# 要绘制的数据
# scores = [50, 55, 60, 65, 70,75,80]
# hours = [1,2,3,4,5,6,7]

x =[]
y = []
for i in range(100):
    tmp =  random.uniform(0,10)
    x.append(tmp)
    tmp2 = 2 * tmp + random.uniform(0,2)
    y.append(tmp2)


# 绘制散点图
plt.scatter(x,y,color = 'blue' ,alpha = 0.5,s=20)


# 添加标题(通用配置)
plt.title('x变量与y变量的关系', color='red', fontsize=20)
# 设置
plt.ylim(0, 20)

# 折线图
plt.plot([0,10], [0,20],color='red',linestyle='--',linewidth=2)

# 自动优化排版
plt.tight_layout()

# 显示图标
plt.show()
```

```python
# 箱线图：数据是否正常，是否有异常数据
'''
在 Jupyter 中绘制箱型图（Box Plot）是探索数据分布、比较分组差异的常用方法。箱型图能直观显示数据的中位数、四分位数范围以及异常值，非常适合快速了解数据集的偏态和离群情况。

1. 箱型图的基本结构
箱子：从下四分位数（Q1，25%）到上四分位数（Q3，75%），中间粗线为中位数（Q2，50%）。
须线（whisker）：通常延伸至 Q1 – 1.5×IQR 和 Q3 + 1.5×IQR 范围内的最远数据点（IQR = Q3 - Q1）。
异常值：超出须线范围的点，通常用圆点或菱形标记。

2. 进阶用法与美化
添加散点叠加：sns.boxplot 搭配 sns.stripplot 或 sns.swarmplot 可显示原始数据分布。

调整颜色：palette 参数控制颜色，例如 palette='Set2'。

水平箱型图：设置 orient='h'（seaborn）或 vert=False（matplotlib）。

自定义须线范围：在 matplotlib 中可通过 whis 参数调整，例如 whis=[5,95] 表示用百分位数确定须线。

3.注意事项
异常值判定：须线外部的点并不一定都是错误数据，需结合业务判断。

样本量：当样本量较小时，箱型图可能无法准确反映分布，建议同时绘制原始点。

多组比较：箱型图适合比较不同组别的分布特征，配合 hue 参数还可实现更细致的分组（如双因素）。

在 Jupyter 中绘制箱型图非常灵活，结合交互式功能（如 %matplotlib notebook）还能实现缩放、平移等操作，方便深入探索数据。

'''
```

```python
# 案列分析：温度分析
# 1。导入库
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rcParams

# 设置字体
rcParams['font.sans-serif'] = ['Heiti TC']
# 2.导入数据
df = pd.read_csv('data/weather.csv')
df.head()

# 3.绘制气温的趋势变化图
df['date']= pd.to_datetime(df['date'])
df.head()

df = df[df['date'].dt.year == 2015]
plt.figure(figsize=(10, 10))
plt.plot(df['date'], df['temp_max'],label='最高气温')
plt.plot(df['date'], df['temp_min'],label='最低气温')
# 配置
plt.title(label= '2015年气温趋势变化图',color='red')
plt.xlabel('日期')
plt.ylabel('温度')
plt.legend(loc='upper right')

```

```python
df['temp_mean'] = (df['temp_max'] + df['temp_min']) /2
df = df[df['date'].dt.year == 2015]
plt.figure(figsize=(10, 10))
plt.plot(df['date'], df['temp_mean'],label='平均气温')
# 配置
plt.title(label= '2015年平均气温趋势变化图',color='red')
plt.xlabel('日期')
plt.ylabel('温度')
plt.legend(loc='upper right')
```
