```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
from matplotlib import rcParams
from tornado.util import errno_from_exception

# 设置字体
rcParams['font.sans-serif'] = ['Heiti TC']
df = pd.read_csv('data/penguins.csv')
df.isna()
df.dropna(inplace=True)
df.head()
# 直方图
sns.histplot(df,x='species') # <Axes: xlabel='species', ylabel='Count'>
```

```python
# 核密度统计图
# <Axes: xlabel='bill_length_mm', ylabel='Density'>
sns.kdeplot(data=df, x='bill_length_mm')
```

```python
sns.histplot(df,x='bill_length_mm',kde=True) # 在直方图的基础上核密度统计
```

```python
'''
计数图用于绘制分类变量的计数分布图，显示每个类别在数据集中出现的次数，是分析分类数据非常直观的工具，可以快速了解类别的分布情况。
#绘制不同岛屿企鹅数量的计数图
'''
# 计数图
sns.countplot(data=df, x="island")
```

```python

'''
散点图
绘制横轴为体重，纵轴为脚蹼长度的散点图。可通过hue参数设置不同组别进行对比。
'''
sns.scatterplot(data=df, x="body_mass_g", y="flipper_length_mm", hue="sex")
```

```python
# 蜂窝图
# 使用 jointplot， kind="hex"
sns.jointplot(data=df, x="body_mass_g", y="flipper_length_mm", kind="hex")
```

```python
# 条形统计图
sns.barplot(data=df, x="species", y="bill_length_mm", estimator='mean',errorbar=None)
```

```python
# 箱线图
sns.boxplot(data=df, x="species", y="bill_length_mm")
```

```python
# 成对关系图

'''成对关系图是一种用于显示多个变量之间关系的可视化工具。它可以展示各个变量之间的成对关系，并且通过不同的图表形式帮助我们理解数据中各个变量之间的相互作用。

对角线上的图通常显示每个变量的分布（如直方图或核密度估计图），帮助观察每个变量的单变量特性。其他位置展示所有变量的两两关系，用散点图表示。'''

sns.pairplot(data=df, hue="species")
```
