```python
# 数据分箱 pd.cut(x,bins,labels)
import pandas as pd

df = pd.read_csv('data/employees.csv')
df.head()
# 对salary进行处理
df1 = df.head(10)[['employee_id', 'salary']]
df1
```

```python
# bins = n ,分成n段区间，起始值、结束值是所有数据的最小值、最大值，用来划分区间
pd.cut(df1['salary'], bins=2)
# 4180~14100~24000
pd.cut(df1['salary'], bins=2).value_counts()
# Categories (3, interval[int64, right]):
# [(0, 10000] < (10000, 20000] < (20000, 30000]]
pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000])  # bins = list ，分成len(list) -1 段
pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000]).value_counts()
# Categories (3, str): ['低' < '中' < '高']
pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000], labels=['低', '中', '高'])
```

```python
# 让每一段的人数相同,等频率分箱
pd.qcut(df1['salary'], 3).value_counts()
```

```python
# 时间数据的处理
import pandas as pd

d = pd.Timestamp('2015-05-02 10:22')  # 2015-05-02 10:22:00
print(d)
print(type(d))  #<class 'pandas.Timestamp'>
print(d.year)
print(d.month)
print(d.day)
print(d.hour)
print(d.minute)
print(d.second)
print('季度：', d.quarter)
print('是否是月底：', d.is_month_end)

# 方法
print('星期几：', d.day_name())
print('转换为天：', d.to_period('D'))  # 转换为天： 2015-05-02
print('转换为季度：', d.to_period('Q'))  # 转换为季度： 2015Q2
print('转换为年度：', d.to_period('Y'))  # 转换为年度： 2015
print('转换为月度：', d.to_period('M'))  # 转换为月度： 2015-05
print('转换为周维度：', d.to_period('W'))  # 转换为周维度： 2015-04-27/2015-05-03
```

```python
# 分组聚合（用来统计）
# df.groupby('分组的字段')['聚合的字段'].聚合函数()
import pandas as pd
df = pd.read_csv('data/employees.csv')
# print(df)
# 先处理缺失值
df['department_id'].isna().sum()
df.dropna(subset=['department_id'], inplace=True)
df['department_id'] = df['department_id'].astype('int64')

# 计算不同部门的平均薪资
df.groupby('department_id').groups # 查看分组
df.groupby('department_id').get_group(10) # 查看具体的某个分组数据
# <pandas.api.typing.SeriesGroupBy object at 0x10a7bb3e0>
# 类似 select sum(salary),department_id from employee group by department_id
# df.groupby('department_id')[['salary']].mean()
df2 = df.groupby('department_id')[['salary']].mean()
df2['salary'] = df2['salary'].round(2)
print(df2)
df2 = df2.reset_index()
df2.sort_values('salary', ascending=False, inplace=True)
```

```python
# 计算不同部门不同岗位的人的平均薪资
df2 = df.groupby(['department_id','job_id'])[['salary']].mean()
df2['salary'] = df2['salary'].round(2)
df2 = df2.reset_index()
df2.sort_values('salary', ascending=False, inplace=True)
print(df2)
```
