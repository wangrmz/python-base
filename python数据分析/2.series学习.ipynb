```python
'''
案例：学生成绩分析
创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间
计算平均分、最高分、最低分，并找出高于平均分的学生人数

'''
# 生成随机成绩
np.random.seed(42)
scores = pd.Series(np.random.randint(50, 101, 10), index=['学生' + str(i) for i in range(1, 11)])
print(scores)
# 平均分
print('平均分：', scores.mean())
print('最高分：', scores.max())
print('最低分：', scores.min())
person = scores[scores > scores.mean()]
# print(scores[scores>scores.mean()])
# print(len(person))
print('高于平均分的学生人数', person.count())
```

```python
'''
案例：温度分析
给定某城市一周每天的最高温度Series，完成以下任务：
1. 找出温度超过30度的天数
2. 计算平均温度
3. 将温度从高到低排序
4. 找出温度变化最大的两天
'''

temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])

# 找出超过30度的天数
days=temperatures[temperatures>temperatures.mean()]
print('温度超过30度的天数:',days.count())

# 计算平均温度
print('平均温度:',temperatures.median())

# 将温度从高到低排序
print('温度从高到低排序:',temperatures.sort_values(ascending=False)) # 默认升序

# 找出温度变化最大的两天
diff = temperatures.diff().abs() # 计算series的变化
print(diff)
print(diff.sort_values(ascending=False).keys()[:2].tolist())

```

```python
'''
案例：股票价格分析
给定某股票连续10个交易日的收盘价Series：
1. 计算每日收益率（当日收盘价/前日收盘价 - 1）
2. 找出收益率最高和最低的日期
3. 计算波动率（收益率的标准差）
'''
# 日期序列
index=pd.date_range('2023-01-01', periods=10)

prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], index=pd.date_range('2023-01-01', periods=10))
print(prices)
# 计算每日收益率,pandas中有转门的函数
prices_daily =prices.pct_change()
print('每日收益率:',prices_daily)
print('收益率最高的日期:',prices_daily.idxmax())
print('收益率最低日期:',prices_daily.idxmin())
print('波动率:',prices_daily.std())
```

```python
'''
案例分析
某产品过去12个月的销售量Series：
1. 计算季度平均销量（每3个月为一个季度）
2. 找出销量最高的月份
3. 计算月环比增长率(当前月和上一个月；同比是和上年的这个月)
4. 找出连续增长超过2个月的月份
'''

sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220], index=pd.date_range('2022-01-01', periods=12, freq='MS'))
print(sales)

# 季度的平均销售量
print('季度平均销量:',sales.resample('QS').mean())  # 重新采样
print('销量最高的月份:',sales.idxmax())
print('月环比增长率:',sales.pct_change())
# 找出连续增长超过2个月的月份
a= sales.pct_change()
b =a>0
# 滑动窗口
b[b.rolling(3).sum()==3].keys().tolist()
```

```python
'''
案例分析：
根据你提供的图片，提取出的文字内容如下：

某商店每小时销售额Series：
1. 按天重采样计算每日总销售额
2. 计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
3. 找出销售额最高的3个小时

'''
# 设置随机数生成器的种子，确保后续生成的随机数序列可重复。
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0, 100, 24),
    index=pd.date_range('2025-01-01', periods=24, freq='h'))

print(hourly_sales)

# 重新采样
daily_sales = hourly_sales.resample('D').sum()
print('天重采样计算每日总销售额:',daily_sales)

# 营业时间/非营业时间
# 计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
# hourly_sales.between_time('8:00','22:00') # 筛选一段时间
# 这里使用布尔索引 获取营业时间的销售额
business_hourly_sales =hourly_sales[(hourly_sales.index.hour>=8) & (hourly_sales.index.hour<=22)]
# 非营业时间销售额 使用drop获取
not_business_hourly_sales = hourly_sales.drop(business_hourly_sales.index)
print('每天营业时间（8:00-22:00）和非营业时间的销售额比例:',business_hourly_sales / (daily_sales - business_hourly_sales))

# 找到销售额最高的三个小时
print('销售额最高的三个小时：',hourly_sales.nlargest(3))

```
